### Histograms

```python
df5.describe()
```

```python
df5['stars'].describe()
```

```python
plt.hist(df5['stars'], bins=25, alpha=0.7)

# Adding titles and labels
plt.title('Histogram of Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')

# Displaying the plot
plt.show()
```

```python
# making sure that zero values in ratings are not actual zeros given by users
df5[(df5['stars'] == 0) & (df5['reviews'] != 0)]
```

```python
df5[df5['reviews'] >= 1356658]
```

```python
df5['reviews'].describe()
```

```python
# Histogram of Reviews with a log scale for y-axis
plt.hist(df5['reviews'], bins=50, color='skyblue', edgecolor='black')
plt.yscale('log')  # Log scale for y-axis
plt.title('Histogram of Reviews (Log Scale)')
plt.xlabel('Reviews')
plt.ylabel('Log Frequency')

plt.show()
```
